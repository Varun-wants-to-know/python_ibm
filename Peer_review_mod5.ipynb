```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import yfinance as yf
```

## Using Webscraping to Extract Stock Data Example


First we must use the `request` library to downlaod the webpage, and extract the text. We will extract Netflix stock data <zzzzz>.


```python
url_gme='https://finance.yahoo.com/quote/GME/history?p=GME'
data_gme=requests.get(url_gme).text
soup_gme = BeautifulSoup(data_gme, 'html5lib')
soup_gme
```

```python
gme_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

for row in soup_gme.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    
    gme_data = gme_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)    
```

We can now print out the dataframe


```python
gme_st=yf.Ticker('GME')
gme=gme_st.history('max')
```

```python
tesla_st=yf.Ticker('TSLA')
tesla=tesla_st.history('max')
```

```python
gme.columns=['Open_GME', 'High_GME', 'Low_GME', 'Close_GME', 'Volume_GME', 'Dividends_GME', 'Stock_Splits_GME']
tesla.columns=['Open_tesla', 'High_tesla', 'Low_tesla', 'Close_tesla', 'Volume_tesla', 'Dividends_tesla', 'Stock_Splits_tesla']
```

```python
df=gme.reset_index().merge(tesla.reset_index(),on='Date',how='outer')
```

```python
df.tail(10)
```

```python
import plotly.express as px

plot=px.line(x='Date',y=['Open_GME','Open_tesla'],data_frame=df,title='Tsla Vs GME',labels={'value':'Stock Value($)','variable':'Stock list'})
newnames = {'Open_GME':'Game_Stop','Open_tesla':'Tesla'}
plot.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
```

```python

plot=px.bar(x='Date',y=['Volume_GME','Volume_tesla'],data_frame=df[(df.Date>'2021-01-01') & (df.Date<'2021-04-01')],title='Tsla Vs GME',labels={'value':'Shares Bought',
'variable':'Stock list'},barmode='group')
newnames = {'Volume_GME':'Game_Stop','Volume_tesla':'Tesla'}
plot.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                    ))       
```
